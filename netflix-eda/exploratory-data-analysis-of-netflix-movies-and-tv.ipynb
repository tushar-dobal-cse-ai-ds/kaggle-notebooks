## 3. Missing Values Analysis

Real-world datasets often contain missing values.

## 4. Data Cleaning

Basic data cleaning is performed by handling missing values in important columns.
This ensures consistency without removing too much data.


```python
df = df.dropna(subset=['type', 'title'])
```

```python
df.isnull().sum()
```

## 5. Data Visualization

Visualizations are used to understand patterns and trends in the Netflix dataset,
such as content type distribution and yearly content growth.


```python
plt.figure(figsize=(7,5))
sns.countplot(data=df,x='type')
plt.title("Count of movies and tv shows on Netflix")
plt.xlabel('type')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x="release_year",order=df['release_year'].value_counts().index[:10])
plt.title("top 10 years with most netflix content")
plt.xticks(rotation=45)
plt.show()
```

## 6. Day 1 Summary

- Netflix has more Movies than TV Shows.
- Content production increased significantly after 2015.
- Certain years show a peak in content addition.
- The dataset contains missing values in columns like country and director,
  which is common in real-world data.


```python
df['country']=df['country'].fillna('Unknown')
df['rating']=df['rating'].fillna('Not Rated')
df['date_added']=df['date_added'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce')
```

```python
df.info()
```

```python
df['year_added']=df['date_added'].dt.year
```

## Feature Engineering

New features were created from existing columns to enable better analysis.
In particular, the year when content was added to Netflix was extracted
from the `date_added` column.


```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='year_added',order=df['year_added'].value_counts().index[:10])
plt.xticks(rotation=45)
plt.title("top years with most content added on netflix")
plt.xlabel('year added')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,y=('rating'),order=df['rating'].value_counts().index)
plt.title('Distribution of content ratings on Netflix')
plt.xlabel('count')
plt.ylabel('ratings')
plt.show()
```

```python
top_countries=df['country'].value_counts().head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=top_countries.values,y=top_countries.index)
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

```python
from collections import Counter
```

```python
genres=df['listed_in'].str.split(',')
all_genres=Counter([genre for sublist in genres.dropna() for genre in sublist])
top_genres=dict(all_genres.most_common(10))
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=list(top_genres.values()), y=list(top_genres.keys()))
plt.title("Top 10 Genres on Netflix")
plt.show()
```
